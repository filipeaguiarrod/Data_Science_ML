Temos mais um dataset para você colocar à prova seus novos conhecimentos.
São 5 datasets do último ano de 5 das maiores empresas de tecnologia do mundo
Apple, Google, Netflix, Amazon e Tesla.
Sua tarefa é adicionar uma coluna com o código de cada ação, passar a data como dtype
datatime e transformá-la em index. Depois deve juntar todos os arquivos em um só.
Quando tiver feito tudo isso, responda a pergunta:
Qual das 5 ações teve a maior média de fechamento no período de julho a setembro de 2019?

```python
import numpy as np
import pandas as pd
```

```python
Apple = pd.read_csv('AAPL.csv')
Apple.head(5)
```

```python
Amazon = pd.read_csv('AMZN.csv')
Google = pd.read_csv('GOOG.csv')
Netflix = pd.read_csv('NFLX.csv')
Tesla = pd.read_csv('TSLA.csv')
```

```python
# Criando coluna 'Code' c/ código das stocks

Apple['Code'] = 'AAPL'

Amazon['Code'] = 'AMZN'

Google['Code'] = 'GOOG'

Netflix['Code'] = 'NFLX'

Tesla['Code'] = 'TSLA'

```

```python
# Datas as datetype

Apple['Date'] = pd.to_datetime(Apple['Date'])

Amazon['Date'] = pd.to_datetime(Amazon['Date'])

Google['Date'] = pd.to_datetime(Google['Date'])

Netflix['Date'] = pd.to_datetime(Netflix['Date'])

Tesla['Date'] = pd.to_datetime(Tesla['Date'])
```

```python
Apple.set_index(['Date'], inplace=True)
Amazon.set_index(['Date'], inplace=True)
Google.set_index(['Date'], inplace=True)
Netflix.set_index(['Date'], inplace=True)
Tesla.set_index(['Date'], inplace=True)

```

```python
#Concatenando todas stocks

Join_stocks = pd.concat([Apple,Amazon,Google,Netflix,Tesla])
Join_stocks.info()
Join_stocks.sort_values('Date')
```

```python
Join_stocks.groupby('Code').mean()
```

```python
result = Join_stocks.loc['2019-09-06':'2019-09-11'].groupby('Code').mean()
result
```
